# Análisis Exploratorio Preliminar — California Housing
Dataset: `sklearn.datasets.fetch_california_housing`

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing

sns.set_theme(style='whitegrid')
```

## 1. Carga de datos

```python
raw = fetch_california_housing(as_frame=True)
df = raw.frame
print(f'Shape: {df.shape}')
df.head()
```

## 2. Información general

```python
df.info()
```

```python
df.describe().round(2)
```

## 3. Valores nulos

```python
nulls = df.isnull().sum()
print('Valores nulos por columna:')
print(nulls[nulls > 0] if nulls.sum() > 0 else 'Sin valores nulos')
```

## 4. Distribución de la variable objetivo (MedHouseVal)

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

df['MedHouseVal'].hist(bins=50, ax=axes[0], color='steelblue')
axes[0].set_title('Distribución MedHouseVal')
axes[0].set_xlabel('Valor mediano (100k USD)')

sns.boxplot(x=df['MedHouseVal'], ax=axes[1], color='steelblue')
axes[1].set_title('Boxplot MedHouseVal')

plt.tight_layout()
plt.show()

print(f'Skewness: {df["MedHouseVal"].skew():.3f}')
print(f'Kurtosis: {df["MedHouseVal"].kurtosis():.3f}')
```

## 5. Distribución de features

```python
features = raw.feature_names
fig, axes = plt.subplots(2, 4, figsize=(16, 8))

for ax, col in zip(axes.flatten(), features):
    df[col].hist(bins=40, ax=ax, color='coral')
    ax.set_title(col)

plt.suptitle('Distribución de features', y=1.02)
plt.tight_layout()
plt.show()
```

## 6. Matriz de correlación

```python
corr = df.corr(numeric_only=True).round(2)

plt.figure(figsize=(10, 8))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='coolwarm',
            center=0, linewidths=0.5)
plt.title('Correlación entre variables')
plt.tight_layout()
plt.show()
```

## 7. Correlación con el target

```python
target_corr = corr['MedHouseVal'].drop('MedHouseVal').sort_values(ascending=False)
target_corr.plot(kind='bar', figsize=(8, 4), color='teal')
plt.title('Correlación de features con MedHouseVal')
plt.axhline(0, color='black', linewidth=0.8)
plt.tight_layout()
plt.show()
print(target_corr)
```

## 8. Scatter: MedInc vs MedHouseVal (feature más correlacionada)

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['MedInc'], df['MedHouseVal'], alpha=0.3, s=5, color='steelblue')
plt.xlabel('Ingreso mediano (MedInc)')
plt.ylabel('Valor mediano casa (MedHouseVal)')
plt.title('MedInc vs MedHouseVal')
plt.tight_layout()
plt.show()
```
